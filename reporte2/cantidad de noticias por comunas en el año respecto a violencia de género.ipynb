```python
import pandas

data1 =  pandas.read_csv('1-4mes/Cantidad de noticias por comuna respecto violencia genero 1-4 meses.csv')
data1.drop('Unnamed: 0', axis=1, inplace=True)


data2 = pandas.read_csv('5-8mes/Cantidad de noticias por comuna respecto violencia genero 5-8 meses.csv')
data2.drop('Unnamed: 0', axis=1, inplace=True)


data3 = pandas.read_csv('9-12mes/Cantidad de noticias por comuna respecto violencia genero 9-12 meses.csv')
data3.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
cantidadNoticiasComunasEnElAñoViolenciaGenero= pandas.concat([data1, data2, data3], axis=0)
```

```python
ciudades = ["Ancud","Castro","Chonchi",
           "Curaco_de_Vélez", "Dalcahue","Puqueldón",
          "Queilén","Quemchi","Quellón",
           "Quinchao","Calbuco","Cochamó",
          "Fresia","Frutillar","Llanquihue",
           "Los_Muermos","Maullín","Puerto_Montt",
          "Puerto_Varas","Osorno","Puerto_Octay",
          "Purranque","Puyehue","Río_Negro",
           "San_Juan_de_la_Costa","San_Pablo","Chaitén",
          "Futaleufú","Hualaihué","Palena"]
```

```python
from pandasql import sqldf 
```

```python
cantidadNoticiasComunasAnoVG = pandas.DataFrame({})
for i in ciudades:
    #print(i)
    result = sqldf(" SELECT SUM("+i+") FROM cantidadNoticiasComunasEnElAñoViolenciaGenero")
    cantidadNoticiasComunasAnoVG[i]=result
cantidadNoticiasComunasAnoVG
```

```python
cantidadNoticiasComunasAnoVG.to_csv('cantidad de noticias por comunas en el año respecto a violencia de género.csv')
```
