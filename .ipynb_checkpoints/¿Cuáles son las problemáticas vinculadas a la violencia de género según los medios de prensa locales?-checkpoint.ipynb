```python
import pandas


dataset1 = pandas.read_csv('data/01-01-21.csv')
dataset1.drop('Unnamed: 0', axis=1, inplace=True)
dataset1
```

```python
import spacy
from spacy.matcher import Matcher
nlp = spacy.load("es_core_news_md")
matcher = Matcher(nlp.vocab)
```

```python
from spacy.matcher import PhraseMatcher
```

```python
cities = ["Ancud", "Castro", "Chonchi",
          "Curaco de Vélez", "Dalcahue", "Puqueldón",
          "Queilén", "Quemchi", "Quellón",
          "Quinchao", "Calbuco", "Cochamó",
          "Fresia", "Frutillar", "Llanquihue",
          "Los Muermos", "Maullín", "Puerto Montt",
          "Puerto Varas", "Osorno", "Puerto Octay",
          "Purranque", "Puyehue", "Río Negro",
          "San Juan de la Costa", "San Pablo", "Chaitén",
          "Futaleufú", "Hualaihué", "Palena", "femicidio", "feminicidio", "acoso", "abuso sexual", "violencia física",
          "violencia sexual", "acoso sexual", "violencia intrafamiliar", "acusado de violación"]
```

```python
for city in cities:
    #print(city)
    dataset1.insert(7,city.replace(" ","_"),0)
```

```python
matcher_cities = PhraseMatcher(nlp.vocab)

for city in cities:
    matcher_cities.add(city, [nlp(city)])
```

```python
for index, row in dataset1.iterrows():
    print(index)
    txt = row["text"]

    try:

        doc = nlp(txt)
        matches_cities = matcher_cities(doc)

        for match_id, start, end in matches_cities:
            span = doc[start:end]  # The matched span
            dataset1.at[index, span.text.replace(" ", "_")] = 1
    except Exception as e:
        print(e)
        pass
```
