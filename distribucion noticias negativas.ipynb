```python
import pandas


dataset = pandas.read_csv('data/dataset_negativas_Quellon')
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset
```

```python
from pandasql import sqldf 
```

```python
fecha9Primeros = []
fechaRestantes = []
fechasOrdenadas = []
for i in range(1, 32):
    if (i < 10):
        # fecha = fecha+str(i)
        # print(fecha)
        result = sqldf(
            "SELECT * from dataset WHERE date == '2021-03-0"+str(i)+"'  ")
        
        fecha9Primeros.append(result.shape[0])
        fechasOrdenadas.append('2021-03-0'+str(i))
    else:
        result = sqldf(
            "SELECT * from dataset WHERE date == '2021-03-"+str(i)+"'  ")
        fechaRestantes.append(result.shape[0])
        fechasOrdenadas.append('2021-03-'+str(i))
```

```python
fechaNegativas = fecha9Primeros+fechaRestantes
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
plt.figure(figsize=(10, 8.5))
plt.bar(fechasOrdenadas, fechaNegativas)
plt.xlabel('Fecha')
plt.ylabel('Nro noticias negativas')
plt.xticks(rotation=90)
plt.show()
plt.savefig('DistibucionNoticiasNegativasMarzoQuellon.png')
```
